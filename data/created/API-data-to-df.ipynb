# Turning csvs into dataframe

```python
import pandas as pd
import re
import glob
```

```python
csv_files = glob.glob('*.csv')

# list to store dfs
dataframes = []

# for filename in current_directory:
#     if filename.endswith('.csv'):
#         print(f"Trying to read file: {filename}")
#         df = pd.read_csv(filename)
#         dataframes.append(df)

for filename in csv_files:        
    print(f"Trying to read file: {filename}")
    df = pd.read_csv(filename)
    dataframes.append(df)
```

```python
len(csv_files)
```

```python
# Concatenate all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
len(combined_df)
```

```python
combined_df.info()
```

```python
combined_df['type'].value_counts()
```

```python
# remove NaN color name values - this should get rid of the color family page.
filtered_df = combined_df.dropna(subset=['color_name'])
filtered_df
```

```python
filtered_df = filtered_df.drop_duplicates(subset=['color_name'])
len(filtered_df)
```

```python
filtered_df[filtered_df['description'].str.contains('Ready-Mixed')]
```

```python
filtered_df = filtered_df.drop(filtered_df[filtered_df['description'].str.contains('Ready-Mixed')].index)
len(filtered_df)
```

```python
filtered_df.reset_index()
```

```python
simple_df = filtered_df[['color_name', 'description']]
simple_df
```

```python
simple_df.info()
```

```python
filtered_df.to_csv('filtered_df.csv', index=False)
```
